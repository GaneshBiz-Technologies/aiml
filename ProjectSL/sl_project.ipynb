```python
# A. Read all the 3 CSV files as DataFrame and store them into 3 separate variables. [1 Mark]
# B. Print Shape and columns of all the 3 DataFrames. [1 Mark]
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# A Custom function for data visualization 
def show_bar_val(ax):
    counts = []
    for bar in ax.patches:
        txval = bar.get_height()
        
        if txval.is_integer()==True:
            txval = int(txval)
        else:
            txval = round(txval, 4)
        
        ax.text( bar.get_x() + bar.get_width() / 2, bar.get_height(), s=txval, ha='center', va='bottom')
        counts.append(bar.get_height())
    return counts

df1_normal = pd.read_csv('Part1+-+Normal.csv')
df1_type_h = pd.read_csv('Part1+-+Type_H.csv')
df1_type_s = pd.read_csv('Part1+-+Type_S.csv')
print("\033[94m Answer B")
print("\033[94m Shape of Normal is {} rows x {} columns".format(df1_normal.shape[0], df1_normal.shape[1]))
print("\033[94m Shape of Type H is {} rows x {} columns".format(df1_type_h.shape[0], df1_type_h.shape[1]))
print("\033[94m Shape of Type S is {} rows x {} columns".format(df1_type_s.shape[0], df1_type_s.shape[1]))
```

```python
# C. Compare Column names of all the 3 DataFrames and clearly write observations. [1 Mark]
nor_cols=df1_normal.columns
h_cols=df1_type_h.columns
s_cols=df1_type_s.columns
diff_count = 0
for v in nor_cols:
    if v in h_cols:
        print('\033[94m Colum {} is same in normal and typeH'.format(v))
    else:
        print('\033[91m Colum {} is diffrent in normal and typeH'.format(v))
        diff_count = diff_count + 1
    if v in s_cols:
        print('\033[94m Colum {} is same in normal and typeS'.format(v))
    else:
        print('\033[91m Colum {} is diffrent in normal and typeS'.format(v))
        diff_count = diff_count + 1

print('-------------------------------------------')
if diff_count == 0:
    print('\033[94m All the column names are found in each other')
else:
    print('\033[91m {} column names are found different'.format(diff_count))
```

```python
'''
Question 2
==========
# Unify all the variations in ‘Class’ feature for all the 3 DataFrames. [1 Marks]
# For Example: ‘tp_s’, ‘Type_S’, ‘type_s’ should be converted to ‘type_s’
'''
df1_normal.replace(to_replace='Normal', value='normal', inplace=True)
df1_normal.replace(to_replace='Nrmal', value='normal', inplace=True)
df1_type_h.replace(to_replace='Type_H', value='type_h', inplace=True)
df1_type_s.replace(to_replace='Type_S', value='type_s', inplace=True)
df1_type_s.replace(to_replace='tp_s', value='type_s', inplace=True)

'''
# B. Combine all the 3 DataFrames to form a single DataFrame [1 Marks]
Checkpoint: Expected Output shape = (310,7)
'''
df1 = pd.merge(left=df1_normal, right=df1_type_h, how='outer')
df1 = pd.merge(left=df1, right=df1_type_s, how='outer')
print('\033[94m Checkpoint: Merged dataframe Rows={} x Columns={}'.format(df1.shape[0], df1.shape[1]))
```

```python
# E. Check 5-point summary of the new DataFrame. [1 Mark]
# Get the  boxplots
fig, axes = plt.subplots(2, 3, sharex=False, sharey=False, figsize=(16, 8))
axes = axes.ravel()
sns.boxplot(data=df1['P_incidence'], ax=axes[0])
sns.boxplot(data=df1['P_tilt'], ax=axes[1])
sns.boxplot(data=df1['L_angle'], ax=axes[2])
sns.boxplot(data=df1['S_slope'], ax=axes[3])
sns.boxplot(data=df1['P_radius'], ax=axes[4])
sns.boxplot(data=df1['S_Degree'], ax=axes[5])
plt.show()
class FivePoint:
  def __init__(self, s):
      self.s= s
      self.mean_ = s.mean()
      self.median_ = s.median()
      self.max_ =s.max()
      self.min_ =s.min()
      self.iqr_ =s.quantile(0.75) - s.quantile(0.25)
      self.lower_q = s.quantile(0.25) + (1.5 * self.iqr_)
      self.upper_q = s.quantile(0.75) + (1.5 * self.iqr_)

P_in_five=FivePoint(df1['P_incidence'])
P_tilt_five=FivePoint(df1['P_tilt'])
L_angle_five=FivePoint(df1['L_angle'])
S_slope_five=FivePoint(df1['S_slope'])
P_radius_five=FivePoint(df1['P_radius'])
S_Degree_five=FivePoint(df1['S_Degree'])

print('\033[94mFive Point summary\n***************************')
print('P_incidence:  Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(P_in_five.mean_, P_in_five.median_, P_in_five.max_,P_in_five.min_, P_in_five.lower_q, P_in_five.upper_q))
print('P_tilt:  Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(P_tilt_five.mean_, P_tilt_five.median_, P_tilt_five.max_,P_tilt_five.min_, P_tilt_five.lower_q, P_tilt_five.upper_q))
print('L_angle:  Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(L_angle_five.mean_, L_angle_five.median_, L_angle_five.max_,L_angle_five.min_, L_angle_five.lower_q, L_angle_five.upper_q))
print('S_slope: Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(S_slope_five.mean_, S_slope_five.median_, S_slope_five.max_,S_slope_five.min_, S_slope_five.lower_q, S_slope_five.upper_q))
print('P_radius: Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(P_radius_five.mean_, P_radius_five.median_, P_radius_five.max_,P_radius_five.min_, P_radius_five.lower_q, P_radius_five.upper_q))
print('S_Degree:  Mean={:2.4f} Median={:2.4f} Max={:2.4f} Min={:2.4f} Lower Quartile={:2.4f} Upper Quartile ={:2.4f}'.\
      format(S_Degree_five.mean_, S_Degree_five.median_, S_Degree_five.max_,S_Degree_five.min_, S_Degree_five.lower_q, S_Degree_five.upper_q))     
```
